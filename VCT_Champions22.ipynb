# Analisando dados de times e jogadores brasileiros

## Apenas jogadores brasileiros

```python
brazil_df = df[df['Country'] == 'Brazil']
brazil_df
```

## Dados dos "Hermanos" que jogam no Brasil (FURIA e LOUD)

```python
arg_df = df[df['Country'] == 'Argentina']
arg_df
```

## Dados dos jogares da FURIA (Organizados por KDA)

```python
furia_df = df[df['Team'] == 'FURIA Esports']
furia_df.sort_values('KDA', ascending = False)
```

```python
plt.bar(furia_df['Player'], furia_df['KDA'])
plt.ylim(1,2)
plt.title('Players da Furia com maior KDA')
plt.xlabel('Players')
plt.ylabel('KDA')
plt.show()
```

```python
plt.bar(furia_df['Player'], furia_df['KD'])
plt.ylim(0.5,1.6)
plt.title('Players da FURIA com maior KD')
plt.xlabel('Players')
plt.ylabel('KD')
plt.show()
```

## Dados dos jogares da LOUD (Organizados por KDA)

```python
loud_df = df[df['Team'] == 'LOUD']
loud_df.sort_values('KDA', ascending = False)
```

```python
plt.bar(loud_df['Player'], loud_df['KDA'])
plt.ylim(1,2)
plt.title('Players da LOUD com maior KDA')
plt.xlabel('Players')
plt.ylabel('KDA')
plt.show()
```

```python
plt.bar(loud_df['Player'], loud_df['KD'])
plt.ylim(0.7,1.6)
plt.title('Players da LOUD com maior KD')
plt.xlabel('Players')
plt.ylabel('KD')
plt.show()
```

# Final LOUD x Optic Gaming

## LOUD:

```python
loud_df
loud_df.sort_values('KDA', ascending = False)
```

### Kills da LOUD

```python
kills_loud = loud_df['K'].sum()
print(kills_loud)
```

### Death da LOUD

```python
death_loud = loud_df['D'].sum()
print(death_loud)
```

### K/D de todo o time da LOUD

```python
kd_loud = kills_loud/death_loud
print("{:.2f}".format(kd_loud))
```

## OpTic Gaming:

```python
optc_df = df[df['Team'] == 'OpTic Gaming']
optc_df.sort_values('KDA', ascending = False)
```

### Deaths da OpTic Gaming
